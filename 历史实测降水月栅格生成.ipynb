```python
import arcpy
arcpy.env.extent = arcpy.Extent(95.877037, 32.160771, 119.065643, 41.860279)
arcpy.env.workspace = r"G:\\day\\yellowriver\\TEM\\year\\"
variable_all = ['pr']
index_all = ["mon\\MON","year"]
for index in index_all:
    for variable in variable_all:
        for year in range(1952, 2011):
            if index == "mon\\MON":
                for mon in range(1, 13):
                    try:
                         out_feature_class = rf"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\shp\\{variable}-{year}-{mon}mean.shp"
                         x_coords = "x"
                         y_coords = "y"
                         z_coords = "z"
                         arcpy.gp.Kriging_sa(out_feature_class, "z",
                                             f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-{mon}-temp.tif",
                                             "Spherical 0.039400", "0.0394",
                                             "VARIABLE 12", "")
                         arcpy.gp.RasterCalculator_sa(f'"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-{mon}-temp.tif" / 10', f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-{mon}-temp2.tif")
                         arcpy.Delete_management(f"G:\\day\\yellowriver\\TEM\\{index}\\chazhi\\tif\\{variable}-{year}-{mon}-temp.tif")
                         arcpy.Clip_management(
                    in_raster= f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-{mon}-temp2.tif",
                    rectangle="95.877037 32.160771 119.065643 41.860279",
                    out_raster=f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-{mon}.tif",
                    in_template_dataset="G:/zhutiansheng/CLIMATE/yellowriver.shp",
                    nodata_value="-3.400000e+038", clipping_geometry="ClippingGeometry",
                    maintain_clipping_extent="NO_MAINTAIN_EXTENT")
                         arcpy.Delete_management(f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-{mon}-temp2.tif")
                         print(f"{variable}-{year}-{mon}.tif-------success!")
                    except Exception as ex:
                        print(f"出现如下异常：{ex}")

            else:
                try:
                    out_feature_class = rf"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\shp\\{variable}-{year}-new.shp"
                    x_coords = "x"
                    y_coords = "y"
                    z_coords = "z"
                    arcpy.gp.Kriging_sa(out_feature_class, "z",
                                        f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-temp.tif",
                                        "Spherical 0.039400", "0.0394",
                                        "VARIABLE 12", "")
                    arcpy.gp.RasterCalculator_sa(f'"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-temp.tif" / 10', f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}.tif")
                    arcpy.Delete_management(f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-temp.tif")
                    arcpy.Clip_management(
                    in_raster= f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}-temp2.tif",
                    rectangle="95.877037 32.160771 119.065643 41.860279",
                    out_raster=f"G:\\day\\yellowriver\\PRE\\{index}\\chazhi\\tif\\{variable}-{year}.tif",
                    in_template_dataset="G:/zhutiansheng/CLIMATE/yellowriver.shp",
                    nodata_value="-3.400000e+038", clipping_geometry="ClippingGeometry",
                    maintain_clipping_extent="NO_MAINTAIN_EXTENT")
                    arcpy.Delete_management(f"G:\\day\\yellowriver\\RRE\\{index}\\chazhi\\tif\\{variable}-{year}-temp2.tif")
                    print(f"{variable}-{year}.tif-------success!")
                except Exception as ex:
                    print(f"出现如下异常：{ex}")

```
