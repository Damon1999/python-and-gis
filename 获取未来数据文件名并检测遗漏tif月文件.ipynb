```python
import os
import re
cFileNumber = 0
model_all = ["MPI-ESM1-2-HR","MRI-ESM2-0"]
variable_all = ["pr","tas","tasmax","tasmin"]
ssp_all = ["ssp126", "ssp245", "ssp370", "ssp585"]
for model in model_all:
    rootdir = 'D:\\futrue\\' + str(model) + "\\"  # 选取删除文件夹的路径,最终结果删除img文件夹 #指定文件夹的路径
    for root, dirs, files in os.walk(rootdir):  # 遍历该文件夹
        for file in files:  # 遍历刚获得的文件名files
            (filename, extension) = os.path.splitext(file)  # 将文件名拆分为文件名与后缀
            if (extension == '.tif'):  # 判断该后缀是否为.c文件
                cFileNumber = cFileNumber + 1  # 记录.c文件的个数为对应文件号
                # print(cFileNumber, os.path.join(root,filename)) #输出文件号以及对应的路径加文件名
                f = open('D:\\futrue\\nc.txt', 'a')
                f.write(str(model) + "-" + filename + "\n")
                f.close()
f = open("D:\\futrue\\nc.txt",'r')
f_line = f.readlines()
for model in model_all:
    for variable in variable_all:
        for ssp in ssp_all:
            for year in range(2015,2101):
                for mon in range(1,13):
                    if mon > 10:
                       filename = str(model) + "-" + str(variable) + "-" + str(ssp) + "-" + str(year) + "-" + str(mon)
                    else:
                        filename = str(model) + "-" + str(variable) + "-" + str(ssp) + "-" + str(year) + "-0" + str(mon)
                    if len(re.findall(filename, str(f_line))) > 0:
                        print("exists!")
                    else:
                        file = open("D:\\futrue\\nc_miss.txt", 'a')
                        file.write(filename + "\n")
                        file.close()
                        print("absent")
```
