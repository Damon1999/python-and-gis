```python
import arcpy
from arcpy.sa import *
arcpy.CheckOutExtension("spatial")
arcpy.gp.overwriteOutput=1
path_workspace = "G:\\zhutiansheng\\nc\\history\\"
variable_all = ["pr","tas","tasmax","tasmin"]
model_all = ["ACCESS-ESM1-5","BCC-CSM2-MR","BCC-ESM1","CanESM5","CNRM-CM6-1","EC-Earth3","GISS-E2-1-G",
             "HadGEM3-GC31-MM","IPSL-CM6A-LR","MIROC6","MPI-ESM1-2-HR","MPI-ESM1-2-LR","MRI-ESM2-0","UKESM1-0-LL"]
D = ".tif"
yellow_river = "yellowriver"
for model in model_all:
    for variable in variable_all:
        arcpy.env.workspace = path_workspace + str(model) + "/" + str(variable) + "/yellowriver/mon/"
        print(arcpy.env.workspace)
        rasters = arcpy.ListRasters()
        for raster in rasters:
            try:
                elevSTDResult = arcpy.GetRasterProperties_management(raster, "MEAN")
                # D:\EC-Earth3\pr\ssp126\tif\yellowriver\year\prssp1262015.tif
                elevSTD = elevSTDResult.getOutput(0)
                content = str(yellow_river) + "," + str(raster) + "," + elevSTD + "\n"
                files = open("G:/zhutiansheng/his_yellowriver_mon.xls", "a")
                files.write(content)
                files.close()
                print("success!")
            except:
                print("false")
                
            
```
