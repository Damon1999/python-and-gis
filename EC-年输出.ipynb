```python
import arcpy
import re
from arcpy.sa import *
arcpy.CheckOutExtension("spatial")
arcpy.gp.overwriteOutput=1
path_workspace = "D:/EC-Earth3/"
variable_all = ["pr","tas","tasmax","tasmin"]
ssp_all = ["ssp126", "ssp245", "ssp370", "ssp585"]
D = ".tif"
yellow_river_all = ["yellowriver","yellowriver1","yellowriver2","yellowriver3"]
for variable in variable_all:
    for ssp in ssp_all:
        for yellow_river in yellow_river_all:
            arcpy.env.workspace = path_workspace + str(variable) + "/" + str(ssp) + "/tif/" + str(yellow_river) + "/year/"
            print(arcpy.env.workspace)
            rasters = arcpy.ListRasters()
            for year in range(2015, 2101):
                for raster in rasters:
                    elevSTDResult = arcpy.GetRasterProperties_management(raster, "MEAN")
                    # D:\EC-Earth3\pr\ssp126\tif\yellowriver\year\prssp1262015.tif
                    elevSTD = elevSTDResult.getOutput(0)
                    content = str(yellow_river) + "," + str(raster) + "," + elevSTD + "\n"
                    files = open("G:/zhutiansheng/ EC_yellowriver_year.xls", "a")
                    files.write(content)
                    files.close()
                    print("success!")
```
