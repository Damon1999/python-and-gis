```python
import arcpy
import re
path_workspace = "D:/EC-Earth3/"
path_temp = "D:/EC-Earth3/temp/"
files = open("D:/EC-Earth3/nc3.txt")
TIF = ".tif" 
NC = ".nc"
for files_no1 in files:
    files_no1 = files_no1.replace("\n", "")
    if len(re.findall("pr", files_no1)) > 0:
        variable = "pr"
    elif len(re.findall("tasmax", files_no1)) > 0:
        variable = "tasmax"
    elif len(re.findall("tasmin", files_no1)) > 0:
        variable = "tasmin"
    elif len(re.findall("tas", files_no1)) > 0:
        variable = "tas"
    ssp1 = re.findall("ssp\d\d\d", files_no1)
    ssp2 = str(ssp1).replace("['", "")
    ssp = str(ssp2).replace("']", "")
    year1 = re.findall("gr_\d\d\d\d", files_no1)
    year2 = str(year1).replace("['gr_", "")
    year = str(year2).replace("']", "")
    for mon in range(1, 13):
        #1,3,5,7,8,10,12 wei0
        if mon in [1,3,5,7,8,10,12]:
            if mon <10:
                B = "time " + "'" + str(year) + "/0" + str(mon) + "/0016" + " 12:00:00'"
            else:
                B = "time " + "'" + str(year) + "/" + str(mon) + "/0016/" + " 12:00:00'"
        #2,4,6,9,11
        else:
            if mon > 10:
                B = "time " + "'" + str(year) + "/" + str(mon) + "/0016" + " 0:00:00'"
            else:
                if mon ==2:
                    B = "time " + "'" + str(year) + "/0" + str(mon) + "/0015" + " 0:00:00'"
                else:
                    B = "time " + "'" + str(year) + "/0" + str(mon) + "/0016" + " 0:00:00'"
        print(B)
        out_raster_layer = str(files_no1) + str(year) + str(mon)
        NC_input = path_workspace + str(variable) + "/" + str(ssp) + "/" + files_no1 + ".nc"
        print(NC_input)
        try:
            out_raster_layer = str(files_no1) + str(year) + str(mon)
            arcpy.MakeNetCDFRasterLayer_md(
                in_netCDF_file=NC_input,
                variable=str(variable), x_dimension="lon", y_dimension="lat",
                out_raster_layer=str(files_no1) + str(year) + str(mon), band_dimension="", dimension_values=B,
                value_selection_method="BY_VALUE")
            arcpy.SelectByDimension_md(in_layer_or_table=str(files_no1) + str(year) + str(mon),
                                       dimension_values=B,
                                       value_selection_method="BY_VALUE")
            try:
                arcpy.CopyRaster_management(in_raster=str(files_no1) + str(year) + str(mon),
                                            out_rasterdataset=path_temp + str(variable) + str(
                                                year) + str(mon) + TIF,
                                            config_keyword="", background_value="",
                                            nodata_value="-3.402823e+038",
                                            onebit_to_eightbit="NONE", colormap_to_RGB="NONE",
                                            pixel_type="",
                                            scale_pixel_value="NONE", RGB_to_Colormap="NONE", format="",
                                            transform="NONE")
                print('copy_finish')
            except Exception as e:
                print(e)
            try:
                arcpy.Resample_management(
                    in_raster=path_temp + str(variable) + str(year) + str(mon) + TIF,
                    out_raster=path_temp + "resample" + str(variable) + str(year) + str(
                        mon) + TIF,
                    cell_size="0.5 0.5",
                    resampling_type="BILINEAR")
                print('resample_finish')
                arcpy.Delete_management(path_temp + str(variable) + str(year) + str(mon) + TIF)
            except Exception as e:
                print(e)
            try:
                arcpy.Clip_management(
                    in_raster=path_temp + "resample" + str(variable) + str(year) + str(
                        mon) + TIF,
                    rectangle="95.8334536376953 32.1461541503906 118.096153637695 41.9296541503906",
                    out_raster=path_workspace +
                               str(variable) + "/" + str(ssp) + "/tif/yellowriver/mon/" + str(
                        variable) + "-" + str(year) + "-" + str(mon) + TIF,
                    in_template_dataset="G:/zhutiansheng/CLIMATE/yellowriver.shp",
                    nodata_value="-3.400000e+038", clipping_geometry="ClippingGeometry",
                    maintain_clipping_extent="NO_MAINTAIN_EXTENT")
                print('success')
            except Exception as e:
                print(e)
            arcpy.Delete_management(path_temp + "resample" + str(variable) + str(year) + str(mon) + TIF)
        except:
            print("pass")
```
