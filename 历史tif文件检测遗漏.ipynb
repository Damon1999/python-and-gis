```python
import os
import re
cFileNumber = 0
model_all = ["ACCESS-ESM1","BCC-CSM2-MR","BCC-ESM1","CanESM5","CNRM-CM6-1","EC-Earth3","GISS-E2-1-G",
             "HadGEM3-GC31-MM","IPSL-CM6A-LR","MIROC6","MPI-ESM1-2-HR","MPI-ESM1-2-LR","MRI-ESM2-0","UKESM1-0-LL"]
variable_all = ["pr","tas","tasmax","tasmin"]
for model in model_all:
    rootdir = 'G:\\zhutiansheng\\nc\\history\\' + str(model) + "\\"  # 选取删除文件夹的路径,最终结果删除img文件夹 #指定文件夹的路径
    for root, dirs, files in os.walk(rootdir):  # 遍历该文件夹
        for file in files:  # 遍历刚获得的文件名files
            (filename, extension) = os.path.splitext(file)  # 将文件名拆分为文件名与后缀
            if (extension == '.tif'):  # 判断该后缀是否为.c文件
                cFileNumber = cFileNumber + 1  # 记录.c文件的个数为对应文件号
                # print(cFileNumber, os.path.join(root,filename)) #输出文件号以及对应的路径加文件名
                f = open('G:\\zhutiansheng\\nc\\history\\tif.txt', 'a')
                f.write(filename + "\n")
                f.close()
                print(filename)  # 以PLACE_RAM(文件名)形式输出文件名
f = open("G:\\zhutiansheng\\nc\history\\tif.txt",'r')
f_line = f.readlines()
for model in model_all:
    for variable in variable_all:
        for year in range(1957,2015):
            for mon in range(1,13):
                if mon <10 :
                    filename = str(model) + "-" + str(variable) + "-" + str(
                        year) + "-0" + str(mon)
                else:
                    filename = str(model) + "-" + str(variable) + "-" + str(
                        year) + "-" + str(mon)
                if len(re.findall(filename, str(f_line))) > 0:
                    print("exists!")
                else:
                    file = open("G:\\zhutiansheng\\nc\history\\tif_miss.txt", 'a')
                    file.write(filename + "\n")
                    file.close()
```
