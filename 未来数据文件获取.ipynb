```python
import os
cFileNumber = 0
model_all = ["MPI-ESM1-2-HR","MRI-ESM2-0"]
for model in model_all:
    rootdir = 'D:\\futrue\\' + str(model) + "\\"  # 选取删除文件夹的路径,最终结果删除img文件夹 #指定文件夹的路径
    for root, dirs, files in os.walk(rootdir):  # 遍历该文件夹
        for file in files:  # 遍历刚获得的文件名files
            (filename, extension) = os.path.splitext(file)  # 将文件名拆分为文件名与后缀
            if (extension == '.tif'):  # 判断该后缀是否为.c文件
                cFileNumber = cFileNumber + 1  # 记录.c文件的个数为对应文件号
                # print(cFileNumber, os.path.join(root,filename)) #输出文件号以及对应的路径加文件名
                f = open('D:\\futrue\\nc1.txt', 'a')
                f.write(filename + "\n")
                f.close()
                print(filename)  # 以PLACE_RAM(文件名)形式输出文件名
```
