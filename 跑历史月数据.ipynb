```python
import arcpy
import re
from netCDF4 import Dataset
import datetime
path_wordspace = "G:/zhutiansheng/nc/history/"
path_temp = "G:/zhutiansheng/nc/history/temp/"
D = ".tif"
out_rasterdataset = "temp1"
files = open("G:/zhutiansheng/nc/history/nc1.txt",'r')
for files_each_line in files:
    #pr_Amon_ACCESS-ESM1-5_historical_r1i1p1f1_gn_185001-201412
    files_each_line = files_each_line.replace('\n', '')
    model = re.findall("_Amon_.*?_historical",files_each_line)
    model = str(model).replace("['_Amon_","")
    model = str(model).replace("_historical']", "")
    variable = re.findall(".*?_Amon",files_each_line)
    variable = str(variable).replace("_Amon']",'')
    variable = str(variable).replace("['", '')
    files_no_target = path_wordspace + str(model) + "/" + files_each_line + ".nc"
    files_no_temp = path_temp
    files_no_final = path_wordspace
    nc = files_no_target
    nc_obj = Dataset(nc)
    info = nc_obj.variables['time'][:]
    time = info.data
    len1 = len(info)
    for i in range(len1):
        timestamp = -3786825600 + (time[i]) * 3600 * 24
        datetime.datetime(1970, 1, 1) + datetime.timedelta(seconds=timestamp)
        info1 = str(datetime.datetime(1970, 1, 1) + datetime.timedelta(seconds=timestamp))
        change = re.findall("-\d\d ",info1)
        change = str(change).replace("['-","")
        change = str(change).replace("']", "")
        info2 = info1.replace(change,"00" + change)
        info2 = info2.replace("-", "/")
        year = re.findall("\d\d\d\d",info1)
        year = str(year).replace("['", "")
        year = str(year).replace("']", "")
        print(int(year))
        if int(year) <1957:
            print("pass")
            continue
        mon = re.findall("-\d\d\-", info1)
        mon = str(mon).replace("['-", "")
        mon = str(mon).replace("-']", "")
        B = "time " + "'" + info2 + "'"
        print(B)
        try:
            arcpy.MakeNetCDFRasterLayer_md(in_netCDF_file= nc, variable=str(variable),
                                           x_dimension="lon", y_dimension="lat",
                                           out_raster_layer=str(files_each_line) + str(year) + str(mon),
                                           band_dimension="", dimension_values=B,
                                           value_selection_method="BY_VALUE")
            arcpy.SelectByDimension_md(in_layer_or_table=str(files_each_line) + str(year) + str(mon),
                                       dimension_values=B,
                                       value_selection_method="BY_VALUE")
        except Exception as e:
            print(e)
        try:
            arcpy.CopyRaster_management(in_raster=str(files_each_line) + str(year) + str(mon),
                                        out_rasterdataset=files_no_temp + str(model) + str(year) + str(mon) + D,
                                        config_keyword="", background_value="",
                                        nodata_value="-3.402823e+038",
                                        onebit_to_eightbit="NONE", colormap_to_RGB="NONE",
                                        pixel_type="",
                                        scale_pixel_value="NONE", RGB_to_Colormap="NONE", format="",
                                        transform="NONE")
            arcpy.Delete_management(str(files_each_line) + str(year) + str(mon))
        except Exception as e:
            print(e)
        try:
            arcpy.Resample_management(
                in_raster=files_no_temp + str(model) + str(year) + str(mon) + D,
                out_raster=files_no_temp + "resample" + str(model) + str(variable) + str(year) + str(mon) + D,
                cell_size="0.5 0.5",
                resampling_type="BILINEAR")
            arcpy.Delete_management(files_no_temp + str(model) + str(year) + str(mon) + D)
        except Exception as e:
            print(e)
        try:
            arcpy.Clip_management(
                in_raster=files_no_temp + "resample" + str(model) + str(variable) + str(year) + str(mon) + D,
                rectangle="95.8334536376953 32.1461541503906 118.096153637695 41.9296541503906",
                out_raster=files_no_final + str(model) + "/" +
                           str(variable) + "/yellowriver/mon/" + str(model) + "-" + str(
                    variable) + "-" + str(year) + "-" + str(mon) + D,
                in_template_dataset="G:/zhutiansheng/CLIMATE/yellowriver.shp",
                nodata_value="-3.400000e+038", clipping_geometry="ClippingGeometry",
                maintain_clipping_extent="NO_MAINTAIN_EXTENT")
            print('success')
            arcpy.Delete_management(
                files_no_temp + "resample" + str(model) + str(variable) + str(year) + str(mon) + D)
        except Exception as e:
            print(e)
```
